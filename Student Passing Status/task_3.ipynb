### Making Extra Column for the passing status
Considering student dataset csv in following link, create another column that represent passing status of a student in maths.
[>33% pass and <33% failed] link: https://www.kaggle.com/datasets/spscientist/students-performance-in-exams?select=StudentsPerformance.csv

```python
import pandas as pd
df = pd.read_csv("StudentsPerformance.csv")
df
```

```python
# since math marks are out of 100
df['Math Status'] = df['math score'].apply(lambda x: 'Pass' if x >= 33 else 'Fail')
```
